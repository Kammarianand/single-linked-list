```python
import pandas as pd

df = pd.read_csv('/content/04-05-2023-TO-04-05-2024-NIFTYBEES-ALL-N (1).csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.columns = [col.strip() for col in df.columns]
```

```python
df.columns
```

```python
# df['Date'] = df['Date'].apply(lambda x: x.strftime("%Y-%m-%d"))
df['Date'] = pd.to_datetime(df['Date'],format='%d-%b-%Y')
```

```python
df.info()
```

```python
cols = ['Total Traded Quantity', 'Turnover ₹', 'No. of Trades',
       'Deliverable Qty']

df[cols] = df[cols].apply(lambda x : x.str.replace(',', '').astype('float'))
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (12,6))
plt.plot(df['Date'], df['Close Price'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid()
plt.show()
```

```python
df.head()
```

```python
import seaborn as sns
plt.figure(figsize = (12, 8))
sns.heatmap(df.select_dtypes(exclude = 'object').corr(), annot = True)
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df['Open Price'] , autolag = 'AIC')
```

```python
print(f"ADF stats = {result[0]}")
```

```python
result
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df['Close Price'], order = (1,1,0))
```

```python
print(model.fit().summary())
```

```python
arima = model.fit()
```

```python
forcast = arima.get_forecast(steps = 10)
```

```python
fro_df = forcast.summary_frame()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12,6))
plt.plot(df['Close Price'])
plt.plot(fro_df['mean'], color = 'red')
plt.fill_between(fro_df.index, fro_df['mean_ci_lower'], fro_df['mean_ci_upper'], color = 'pink')
```

```python
fro_df.head()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

sar_model = SARIMAX(df['Close Price'], order = (1,1,1), seasonal_order = (0,0,0,0))
```

```python
print(sar_model.fit().summary())
```

```python
model = sar_model.fit()

forcast = model.get_forecast(steps = 20)
s_df = forcast.summary_frame()
```

```python
s_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12,6))
plt.plot(df['Close Price'])
plt.plot(s_df['mean'], color = 'red')
plt.fill_between(s_df.index, s_df['mean_ci_lower'], s_df['mean_ci_upper'], color = 'pink')
```
