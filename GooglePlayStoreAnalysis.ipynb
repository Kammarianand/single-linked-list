<strong>Data Preprocesssing</strong>

```python
#checking for null vales
dataFrame.isnull().sum()
```

```python
#droping null values
df = dataFrame.dropna()
```

```python
df.isnull().sum()
```

```python
#data analysis on rating and review columns

#finding average rating

avg_rating = int(sum(df['Rating']))//len(df['Rating'])

print("Average rating of apps: ",avg_rating)
```

```python
#apps with more than 4 rating
apps_rating = df['Rating'].gt(4).sum()

print(f"{apps_rating} apps with rating > 4")
```

```python
type(df['Reviews'])
```

```python
r = 0
for i in df['Rating']:
  r = r + 1
r
print(f"total reviews {r}")
```

<p>Categorical data</p>

```python
#total no.of categories
print(f"total no.of categories {len(df['Category'].unique())}")
```

```python
#total no.of apps in ART_AND_DESIGN category
sum = 0
for i in df['Category']:
  if i == 'ART_AND_DESIGN':
    sum = sum+1
sum
```

```python
df['Type'].unique()
```

```python
f = 0
for i in df['Type']:
  if i == 'Free':
    f = f+1
print("Total no.of free applications: ",f)

```

```python
#percentage of free apps in the dataset
print(f"{int(f/len(df['Type'])*100)}% are free")
```

```python
#content Rating
for i in df['Content Rating'].unique():
  print(i)
```

<p> Total no.of apps in each category</p>

```python
for name in df['Category'].unique():
  c = 0
  for i in df['Category']:
    if i == name:
      c = c+1
  print(name,":",c)
```

```python
#total no.of apps in content rating
content_rating = {}
for name in df['Content Rating'].unique():
  ct = 0
  for i in df['Content Rating']:
    if i == name:
      ct = ct+1
  content_rating[name] = ct
print(content_rating)
```

<strong>using imputer</strong>

```python
from sklearn.impute import SimpleImputer
```

```python
simputImpute = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
dF['Salary']
```

```python
simputImpute.fit(dF.iloc[:,1:3].values)

data = simputImpute.transform(dF.iloc[:,1:3].values)
```

```python
dF.iloc[:,1:3] = data
```

```python
impute = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
impute.fit(dF.iloc[:,:].values)

dF.iloc[:,:] = impute.transform(dF.iloc[:,:].values)
```
